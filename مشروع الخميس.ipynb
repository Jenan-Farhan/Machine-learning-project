```python
# Import some necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
# Import data
# Look at top 5 rows with .head()
data = pd.read_csv("/content/car_prediction_data.csv")
data.head()
```

```python
# Change the types of some columns
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

data['Fuel_Type'] = le.fit_transform(data['Fuel_Type'])

data
```

```python
# Change the types of some columns
le = LabelEncoder()

data['Seller_Type'] = le.fit_transform(data['Seller_Type'])

data
```

```python
# Change the types of some columns
le = LabelEncoder()

data['Transmission'] = le.fit_transform(data['Transmission'])

data
```

```python
# Change the types of some columns
le = LabelEncoder()

data['Car_Name'] = le.fit_transform(data['Car_Name'])

data
```

```python
# use .isna().sum() on df to get missing counts for all columns
data.isnull().sum()
```

```python
# getting the statistical measures of the data
data.describe()
```

```python
# Distribution of price column values
sns.set_style("whitegrid")
plt.figure(figsize=(15, 10))
sns.distplot(data.Selling_Price)
plt.show()
```

```python
# Correlation between all data features
data.corr()
```

```python
plt.figure(figsize=(20, 15))
correlations = data.corr()
sns.heatmap(correlations, cmap="coolwarm", annot=True)
plt.show()
```

```python
# separating the data and labels
X = data.drop('Selling_Price', axis=1)
y = data['Selling_Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

**DecisionTreeRegressor**

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error


model = DecisionTreeRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```

#Calculate the **score**

```python
model.score(X_test,y_test)
```
